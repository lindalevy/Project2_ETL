Take two new csv files, ATO clean and regional postcodes scrapped from web, and merge them 
Partially needed as regional postcodes included 'postcodes' in the postcode range that were not actually valid
postcodes. Also benefical as this is a one to one relationship, each csv file
contianing one occurance of the postcode

```python
# import libraries
import pandas as pd
```

```python
#read in two postcode files
df1 = pd.read_csv('..\Resources\Input_Files\ATO_clean.csv')
df2 = pd.read_csv('..\Resources\Input_Files\Web_Scraping_postcode.csv')
```

```python
df1
```

```python
df2
```

```python
# merge the two dataframes on postcode, left join to only include post codes in the ATO_clean file
# This is because the ATO file are all genuine post codes, and it is only for these that we
# need the addditional data
output=pd.merge(left =df1,right=df2,how="left", on="Postcode")
output
```

```python
#because it is merged with a regional postcode file, the area as metro  
#needs to be set on all NaN postcode values
output['Area'].fillna('Metro', inplace=True)

final_output= output.drop_duplicates()
final_output
```

```python
#write dataframe to csv
output.to_csv("../Resources/Output_Files/ATO_Area_Clean.csv",index=False)
```
