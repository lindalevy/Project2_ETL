Looking here to collect a suburb name from this existing csv file that was downloaded

```python
# import libraries
import math
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None)
```

```python
#Read in csv file
path="../Resources/Input_Files/australian_postcodes.csv"
APC_df = pd.read_csv(path, low_memory=False)
```

```python
# Show DataFrame
APC_df
```

```python
# Get the column names so that necessary column names can be extracted.
APC_df.columns
```

```python
# Select necessary columns and save in new DataFrame
APC1_df = APC_df.iloc[:,[1,2,3]]
APC1_df
```

```python
#Rename Suburb column
APC_clean_df=APC1_df.rename(columns={APC1_df.columns[1]: 'Suburb'})
APC_clean_df
```

```python
# Getting data for Victoria only
VIC_df = APC_clean_df[APC_clean_df['state']=='VIC']
VIC_df
```

```python
#remove rows and columns with Null/NaN values
VIC_df.dropna()
VIC_df
```

```python
# Drop duplicates
VIC_df = VIC_df.drop_duplicates(subset='postcode', keep="first")
VIC_df
```

```python
# Check data types
VIC_df.dtypes
```

```python
#write dataframe to csv
VIC_df.to_csv("../Resources/Output_Files/postcode_clean.csv", index =False)
```
