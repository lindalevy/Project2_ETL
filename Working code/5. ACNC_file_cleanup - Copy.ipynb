Clean up and assess the csv file from ACNC Registered Charities: https://data.gov.au/data/dataset/b050b242-4487-4306-abf5-07ca073e5594

```python
# import libraries
import math
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None)
```

```python
path="../Resources/Input_Files/ACNC_registered_charities.csv"
ACNC_df = pd.read_csv(path, low_memory=False)
```

```python
# Get the column names so that necessary column names can be extracted.
ACNC_df.columns
```

```python
# copy ABN number into a second column for extraction purposes
#this will be needed becuase I extract in two parts and then then to merge back to one DF
ACNC_df["ABN_Copy"] = ACNC_df["ABN"]
ACNC_df
```

```python
#Extract the needed columns in two parts, so that I can use a range for part 2
#Each part has the ABN  number in it
df1 = ACNC_df.iloc[:,[0,1,8,9,14,23,26]]
df2 = ACNC_df.iloc[:,28:67]
df2_updated = df2.rename(columns={'ABN_Copy': 'ABN'})

#Merge dataframes
output=pd.merge(left =df1,right=df2_updated,how="inner", on="ABN")
output
```

```python
#As it is a link, check postcode for valid entries
#check for invalid postcodes
output['Postcode'].isnull().values.any()
```

```python

#some NaN values in postcode, so replace those
output['Postcode'] = output['Postcode'].fillna(0)

output['Postcode'].isnull().values.any()
```

```python
# converting postcode from float to int 
output.dtypes
output['Postcode'] = output['Postcode'].apply(np.int64)
output
```

```python
#Vic, VIC and VIctoria need standardising
output['State'].mask(output['State'] == 'Victoria', 'VIC')
output
```

```python
# Keeping data for Victoria only
VIC_df = output[output['State']=='VIC']
VIC_df
```

```python
#there were a lot of Nan values. Looking at the source file, these should have been blank
#As NaN isnot providing useful data, replace them all with blanks

VIC_clean_df = VIC_df.replace(np.nan, '', regex=True)
VIC_clean_df
```

```python
VIC_clean_df.dtypes
```

```python
#write dataframe to csv as a potential output file
#
VIC_clean_df.to_csv("../Resources/Output_Files/ACNC_clean.csv", index =False)
```
